```python
from rdkit.Chem import AllChem as chem

caffeine = chem.MolFromSmiles('Cn1cnc2c1c(=O)n(C)c(=O)n2C')
caffeine
```

```python
theobromine = chem.MolFromSmiles('Cn1cnc2c1c(=O)[nH]c(=O)n2C')
theobromine
```

```python
propane = chem.MolFromSmiles('CCC')
propane
```

```python
from rdkit.Chem import AllChem as chem
import rdkit.DataStructs as data

morgan = chem.GetMorganGenerator(radius=3, fpSize=512)
caffeine_fp = morgan.GetFingerprint(caffeine)
theobromine_fp = morgan.GetFingerprint(theobromine)
data.DiceSimilarity(caffeine_fp, theobromine_fp)
```

```python
propane_fp = morgan.GetFingerprint(propane)
data.DiceSimilarity(caffeine_fp, propane_fp)
```
